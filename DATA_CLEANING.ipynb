# DATA CLEANSING

```python
#https://www.youtube.com/watch?v=0irmDBWLrco
import pandas as pd
import numpy as np 
```

```python
PL21 = pd.read_csv('PL21.csv')
PL22 = pd.read_csv('PL22.csv')
PL23 = pd.read_csv('PL23.csv')
PL24 = pd.read_csv('PL24.csv')
PL = pd.concat([PL21, PL22, PL23, PL24], axis = 0)
PL = PL[PL.COMP=='Premier League']
PL.shape
```

```python
PL.TEAM.value_counts()
```

```python
PL.REFEREE.value_counts()
```

```python
PL.ROUND.value_counts()
```

```python
PL.head()

```

```python
PL.loc[:,'YEAR'] = PL.DATE.str[0:4].astype(int)
PL.loc[:,'DATES'] = pd.to_datetime(PL['DATE'])
```

```python
PL.dtypes
```

```python
PL.loc[:,'VENUE_CD'] = PL.VENUE.astype("category").cat.codes
PL
```

```python
PL.loc[:,'VOPPONENT_CD'] = PL.OPPONENT.astype("category").cat.codes
PL
```

```python
PL.loc[:,'HOUR'] = PL.TIME.str.replace(r":.+", "", regex = True).astype(int)
PL
```

```python
PL.loc[:,'DAY_CD'] = PL.DATES.dt.dayofweek
PL
```

```python
PL.loc[:,'TARGET'] = (PL.RESULT == 'W').astype(int)
PL
```

```python
PL.to_csv('PL.csv', index = False)
```
